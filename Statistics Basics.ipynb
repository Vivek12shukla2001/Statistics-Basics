#Statistics Basics

1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.


Qualitative Data (Categorical): Describes characteristics or qualities.

Examples: Eye color, nationality.

Nominal Scale: Categories with no order (e.g., blood type).

Ordinal Scale: Categories with a meaningful order but no uniform gap (e.g., class rank).

Quantitative Data (Numerical): Represents measurable quantities.

Examples: Age, height, temperature.

Interval Scale: Numeric with equal intervals, but no true zero (e.g., temperature in Celsius).

Ratio Scale: Numeric with a true zero (e.g., weight, height, age).

2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.


Mean: Average; best for symmetrical distributions.

Example: Mean of [2, 4, 6] = 4

Median: Middle value; best for skewed data or when outliers exist.

Example: Median of [1, 3, 100] = 3

Mode: Most frequent value; best for categorical or multimodal data.

Example: Mode of [1, 2, 2, 3] = 2

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?


Dispersion: Measures the spread of data values from the central point.

Variance: Average of squared deviations from the mean.

Standard Deviation (SD): Square root of variance; indicates how much values typically differ from the mean.

Higher SD or variance = more variability.

4. What is a box plot, and what can it tell you about the distribution of data?


A box plot shows the five-number summary: Min, Q1, Median (Q2), Q3, Max.

It reveals:

Spread of the data

Skewness (symmetry/asymmetry)

Outliers (data points outside 1.5Ã—IQR from Q1 or Q3)

5. Discuss the role of random sampling in making inferences about populations.


Random sampling ensures every member of a population has an equal chance of selection.

It helps:

Reduce bias

Increase generalizability

Make valid statistical inferences

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?


Skewness measures the asymmetry of a distribution.

Positive skew (Right): Tail on the right, mean > median.

Negative skew (Left): Tail on the left, mean < median.

Zero skew: Symmetrical distribution.

Effect: Skewness influences which measure of central tendency is most accurate (median is preferred for skewed data).

7. What is the interquartile range (IQR), and how is it used to detect outliers?


IQR = Q3 - Q1

Represents the middle 50% of the data.

Outlier Detection Rule:

Below: Q1 - 1.5Ã—IQR

Above: Q3 + 1.5Ã—IQR

Outliers fall outside this range.

8. Discuss the conditions under which the binomial distribution is used.


Use binomial distribution when:

Fixed number of trials (n)

Two outcomes (success/failure)

Constant probability of success (p)

Independent trials

Example: Tossing a coin 5 times.

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).


Normal Distribution Properties:

Bell-shaped, symmetric

Mean = Median = Mode

Defined by mean and standard deviation

Empirical Rule:

68% of data within 1 SD of mean

95% within 2 SDs

99.7% within 3 SDs
